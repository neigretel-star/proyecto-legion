```python
import pandas as pd
# with open (...)
df = pd.read_csv("../data/clean/weather_data.csv")
df.head(1)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
temperatura = df["temperature"]

plt.figure(figsize=(10,6))
temperatura.plot(kind="line")

plt.title("Temperatura en Madrid Barajas")
plt.xlabel("Registro")
plt.ylabel("Temperatura (°C)")

plt.show()
```

```python
viento = df["wind_speed"]

plt.figure(figsize=(10,6))
viento.plot(kind="line")

plt.title("Velocidad del viento en Madrid Barajas")
plt.xlabel("Registro")
plt.ylabel("Velocidad viento")

plt.show()
```
